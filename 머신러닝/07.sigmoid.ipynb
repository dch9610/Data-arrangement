$$
{f(x)} = {1 \over 1+e^{-x}}
$$

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
print( 2**2)
print( 2**3)
print( 2**-2)
```

```python
# sigmoid 함수 만들기
def sigmoid(z):
    return 1/ (1+math.e ** -z)

print(sigmoid(-10))
print(sigmoid(-1))
print(sigmoid(0))
print(sigmoid(1))
print(sigmoid(10))
```

```python
for z in np.linspace(-10,10,50):
    s= sigmoid(z)
    plt.plot(z, s,'ro')
plt.show()
```
